```python
df = pd.read_csv('aaa.txt')
```

```python
p = df.pivot_table(index=['time'],columns=['dt'],values=['value1','value2','value3']).reorder_levels([1,0], axis=1)
```

```python
p
```

```python
# 通过写死的方式实现需求
p.iloc[:,[0,3,6,1,4,7,2,5,8]]
```

### 采用优雅的写法

```python
# 获取全部列数
column_count = len(p.columns.labels[0])
# 获取二级分类的个数
second_index_count = len(p.columns.levels[1])
# 生成新的索引
column_indexes = sorted(range(column_count),key=lambda x:x%second_index_count)
column_indexes
```

```python
p.iloc[:,column_indexes]
```
